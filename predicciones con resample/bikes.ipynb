Primer Prueba. Quitamos columnas que no vamos a utilizar, agrupamos fecha y valores medios

filtro por horas con mas utilización para poderar las horas de mayor cantidad de renta

```python
df_agrup_fecha = df_train[df_train['hr'] > 5] 
df_agrup_fecha = df_agrup_fecha.rename(columns= {'dteday' : 'ds', 'cnt' : 'y'})
```

```python
df_agrup_fecha = df_agrup_fecha[['ds', 'y', 'season', 'holiday', 'weekday', 'workingday', 'weathersit', 'temp', 'atemp', 'hum', 'windspeed']]
df_agrup_fecha = df_agrup_fecha.groupby('ds').mean()
df_agrup_fecha = df_agrup_fecha.reset_index('ds')
```

```python
df_agrup_fecha['clima'] = round(df_agrup_fecha['weathersit'])
```

```python
df_agrup_fecha.drop(columns= 'weathersit', inplace= True)
```

```python
df_agrup_fecha = df_agrup_fecha.set_index('ds')
```

```python
from prophet import Prophet
```

```python
df_agrup_fecha = df_agrup_fecha.reset_index()
df_agrup_fecha.head(-5)
```

```python
modelo2 = Prophet()
modelo2.add_regressor('season')
modelo2.add_regressor('holiday')
modelo2.add_regressor('weekday')
modelo2.add_regressor('clima')
modelo2.add_regressor('workingday')
modelo2.add_regressor('temp')
modelo2.add_regressor('atemp')
modelo2.add_regressor('hum')
modelo2.add_regressor('windspeed')
modelo2.fit(df_agrup_fecha)
```

```python
df_agrup_fecha = df_agrup_fecha.set_index('ds')
df_agrup_fecha.head()
```

```python
df_test_agrup_fecha = df_test[df_test['hr'] > 5]
df_test_agrup_fecha = df_test_agrup_fecha.rename(columns= {'dteday' : 'ds'}) 
```

```python
df_test_agrup_fecha = df_test_agrup_fecha[['ds', 'season', 'holiday', 'weekday', 'workingday', 'weathersit', 'temp', 'atemp', 'hum', 'windspeed']]
df_test_agrup_fecha = df_test_agrup_fecha.groupby('ds').mean()
df_test_agrup_fecha = df_test_agrup_fecha.reset_index('ds')
```

```python
df_test_agrup_fecha['clima'] = round(df_test_agrup_fecha['weathersit'])
df_test_agrup_fecha.drop(columns= 'weathersit', inplace= True)
df_test_agrup_fecha = df_test_agrup_fecha.set_index('ds')
```

```python
df_test_agrup_fecha.head()
```

```python
def estacion(ds):
    date = (pd.to_datetime(ds)).date()
    
    if df_agrup_fecha[date:].empty:
        return df_test_agrup_fecha[date:]['season'].values[0]
    else:
        return (df_agrup_fecha[date:]['season']).values[0]
    
    return 0

def clima(ds):
    date = (pd.to_datetime(ds)).date()
    
    if df_agrup_fecha[date:].empty:
        return df_test_agrup_fecha[date:]['clima'].values[0]
    else:
        return (df_agrup_fecha[date:]['clima']).values[0]
    
    return 0

def temp(ds):
    date = (pd.to_datetime(ds)).date()
    
    if df_agrup_fecha[date:].empty:
        return df_test_agrup_fecha[date:]['temp'].values[0]
    else:
        return (df_agrup_fecha[date:]['temp']).values[0]
    
    return 0

def sen_tem(ds):
    date = (pd.to_datetime(ds)).date()
    
    if df_agrup_fecha[date:].empty:
        return df_test_agrup_fecha[date:]['atemp'].values[0]
    else:
        return (df_agrup_fecha[date:]['atemp']).values[0]
    
    return 0

def vacaciones(ds):
    date = (pd.to_datetime(ds)).date()
    
    if df_agrup_fecha[date:].empty:
        return df_test_agrup_fecha[date:]['holiday'].values[0]
    else:
        return (df_agrup_fecha[date:]['holiday']).values[0]
    
    return 0

def dia_semana(ds):
    date = (pd.to_datetime(ds)).date()
    
    if df_agrup_fecha[date:].empty:
        return df_test_agrup_fecha[date:]['weekday'].values[0]
    else:
        return (df_agrup_fecha[date:]['weekday']).values[0]
    
    return 0

def feriado(ds):
    date = (pd.to_datetime(ds)).date()
    
    if df_agrup_fecha[date:].empty:
        return df_test_agrup_fecha[date:]['workingday'].values[0]
    else:
        return (df_agrup_fecha[date:]['workingday']).values[0]
    
    return 0

def humedad(ds):
    date = (pd.to_datetime(ds)).date()
    
    if df_agrup_fecha[date:].empty:
        return df_test_agrup_fecha[date:]['hum'].values[0]
    else:
        return (df_agrup_fecha[date:]['hum']).values[0]
    
    return 0

def viento(ds):
    date = (pd.to_datetime(ds)).date()
    
    if df_agrup_fecha[date:].empty:
        return df_test_agrup_fecha[date:]['windspeed'].values[0]
    else:
        return (df_agrup_fecha[date:]['windspeed']).values[0]
    
    return 0

```

```python
future = modelo2.make_future_dataframe(periods= len(df_test_agrup_fecha))
future['season'] = future['ds'].apply(estacion)
future['holiday'] = future['ds'].apply(vacaciones)
future['weekday'] = future['ds'].apply(dia_semana)
future['clima'] = future['ds'].apply(clima)
future['workingday'] = future['ds'].apply(feriado)
future['temp'] = future['ds'].apply(temp)
future['atemp'] = future['ds'].apply(sen_tem)
future['hum'] = future['ds'].apply(humedad)
future['windspeed'] = future['ds'].apply(viento)
```

```python

forecast2 = modelo2.predict(future)
```
